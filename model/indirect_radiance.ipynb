# Indirect radiance


```python
import matplotlib.pyplot as plt
from utils import read_ground_truth_zip
```

```python
GROUND_TRUTH_FILE = '../data/radiance/chameleon_front.zip'
arrays, parameters = read_ground_truth_zip(GROUND_TRUTH_FILE, verbose=True)
```

```python
index = 0

frame = arrays[index]
height = parameters['resolution']
radiance = frame[:, 5:].reshape(height, height, 3)

plt.figure(figsize=(6, 6))
plt.imshow(radiance)
plt.axis('off')
plt.show()
```
