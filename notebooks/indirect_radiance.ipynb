# Indirect radiance


```python
from math import sqrt
import matplotlib.pyplot as plt
from utils import read_ground_truth_zip
```

```python
GROUND_TRUTH_FILE = '../data/radiance/ground_truth.zip'
arrays = read_ground_truth_zip(GROUND_TRUTH_FILE, verbose=True)
```

```python
index = 0

# Each frame contains 8 values per pixel (position, azimuth, elevation, radiance)
frame = arrays[index]
pixels = frame.shape[0]

height = int(sqrt(pixels))
radiance = frame[:, 5:].reshape(height, height, 3)

plt.figure(figsize=(6, 6))
plt.imshow(radiance)
plt.axis('off')
plt.show()
```
